# Google stock analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_raw=pd.read_csv("data/GOOGL_5yr.csv")
```

```python
df_raw
```

```python
df_raw["Date"]=pd.to_datetime(df_raw["Date"])
```

```python
df_raw
```

```python
df_raw.dtypes
```

```python
df_new=df_raw.set_index("Date")
df_new.head(3), df_new.tail(3)
df_new.loc["2020-04-27":"2021-04-27"]
```

```python
fig,ax = plt.subplots()
plt.plot(df_new["Close"])

#plt.plot(df_new["Open"])
ax.set_xlabel("Date")
ax.set_ylabel("Closing price (USD)")
plt.legend()
plt.title("Close values")
```

```python
s_mean_10 = df_new["Close"].rolling(10).mean()
s_mean_10

s_mean_50 = df_new["Close"].rolling(50).mean()
s_mean_50

s_mean_25 = df_new["Close"].rolling(25).mean()
s_mean_25
```

# Moving average plots

```python
fig,ax = plt.subplots()
plt.figure(figsize=(20,8))
ax.set_xlabel("Date")
ax.set_ylabel("Closing price (USD)")
plt.title("Close values")

plt.plot(df_new["Close"])
plt.plot(s_mean_10, label="Rolling 10")
plt.plot(s_mean_25, label="Rolling 25", linestyle="dotted")
plt.plot(s_mean_50, label="Rolling 50", linestyle="dashed")
plt.plot(df_new["Close"].rolling(100).mean(), label="Rolling 100", linestyle="solid"  )
plt.plot(df_new["Close"].rolling(500).mean(), label="Rolling 500", linestyle="solid"  )
plt.legend()


#plt.plot(df_new["Open"])
```
