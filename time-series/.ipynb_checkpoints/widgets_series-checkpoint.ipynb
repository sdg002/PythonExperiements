# Widget time series

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_raw=pd.read_csv("data/ch04_widget_sales.csv")
```

```python
df_raw
```

# Raw plot

```python
fig,ax = plt.subplots()
plt.figure(figsize=(20,8))
ax.set_xlabel("Date")
ax.set_ylabel("Closing price (USD)")
plt.title("Widget sales")

plt.plot(df_raw["widget_sales"])
plt.legend()
```

# Rolling mean plots

```python
fig,ax = plt.subplots()
plt.figure(figsize=(20,8))
ax.set_xlabel("Date")
ax.set_ylabel("Closing price (USD)")
plt.title("Widget sales")

plt.plot(df_raw["widget_sales"])
plt.plot(df_raw["widget_sales"].rolling(10).mean(), label="Rolling 10", linestyle="solid"  )
plt.plot(df_raw["widget_sales"].rolling(5).mean(), label="Rolling 5", linestyle="dotted"  )
plt.legend()
```
