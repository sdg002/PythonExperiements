## Forecasting functions
### Forecast using mean

```python
def rolling_forecast_mean(df: pd.DataFrame, train_len: int, horizon: int, window: int) -> list:
    total_len = train_len + horizon
    end_idx = train_len
    pred_mean = []
    
    for i in range(train_len, total_len, window):
        mean = np.mean(df[:i].values)
        pred_mean.extend(mean for _ in range(window))
        
    return pred_mean

```

### Forecast using last value

```python
def rolling_forecast_last(df: pd.DataFrame, train_len: int, horizon: int, window: int) -> list:
    total_len = train_len + horizon
    end_idx = train_len

    pred_last_value = []
    
    for i in range(train_len, total_len, window):
        last_value = df[:i].iloc[-1].values[0]
        pred_last_value.extend(last_value for _ in range(window))
        
    return pred_last_value
```

### Forecast using AR

```python
def rolling_forecast_ar(df: pd.DataFrame, train_len: int, horizon: int, window: int, lags:int) -> list:
    total_len = train_len + horizon
    end_idx = train_len

    pred_AR = []
    
    for i in range(train_len, total_len, window):
        model = SARIMAX(df[:i], order=(lags,0,0))
        res = model.fit(disp=False)
        predictions = res.get_prediction(0, i + window - 1)
        oos_pred = predictions.predicted_mean.iloc[-window:]
        pred_AR.extend(oos_pred)
        
    return pred_AR
```

## Do the predictions

### Mean

```python
pred_mean=rolling_forecast_mean(df_diff, TRAIN_LEN ,HORIZON,WINDOW)
len(pred_mean), pred_mean[:10]
```

### Last Value

```python
pred_last_value=rolling_forecast_last(df=df_diff, train_len=TRAIN_LEN, horizon=HORIZON, window=WINDOW)
len(pred_last_value), pred_last_value[:10]
```

### AR lags=2

```python
pred_AR_2=rolling_forecast_ar(df=df_diff, train_len=TRAIN_LEN, horizon=HORIZON, window=WINDOW, lags=2)
len(pred_AR_2), pred_AR_2[:10]
```

### AR lages=3

```python
pred_AR_3=rolling_forecast_ar(df=df_diff, train_len=TRAIN_LEN, horizon=HORIZON, window=WINDOW, lags=3)
len(pred_AR_3), pred_AR_3[:10]
```

```python
test['pred_mean']=pred_mean
test['pred_last_value']=pred_last_value
test['pred_AR_2']=pred_AR_2
test['pred_AR_3']=pred_AR_3
test.head(3)
```

## Plot of various predictions

```python
fig, ax = plt.subplots()

ax.plot(df_diff['foot_traffic_diff'])
ax.plot(test['foot_traffic_diff'], 'b-', label='actual')
ax.plot(test['pred_mean'], 'g:', label='mean')
ax.plot(test['pred_last_value'], 'r-.', label='last')
ax.plot(test['pred_AR_2'], 'k.', label='AR(2)')
ax.plot(test['pred_AR_3'], 'k--', label='AR(3)')

ax.legend(loc=2)

ax.set_xlabel('Time')
ax.set_ylabel('Diff. avg. weekly foot traffic')

ax.axvspan(947, 998, color='#808080', alpha=0.2)

ax.set_xlim(920, 999)

plt.xticks([936, 988],[2018, 2019])

fig.autofmt_xdate()
plt.tight_layout()

plt.savefig('figures/CH05_F10_peixeiro.png', dpi=300)
```

```python
from sklearn.metrics import mean_squared_error

mse_mean = mean_squared_error(test['foot_traffic_diff'], test['pred_mean'])
mse_last = mean_squared_error(test['foot_traffic_diff'], test['pred_last_value'])
mse_AR_2 = mean_squared_error(test['foot_traffic_diff'], test['pred_AR_2'])
mse_AR_3 = mean_squared_error(test['foot_traffic_diff'], test['pred_AR_3'])

print(f"{mse_mean=}, {mse_last=}, {mse_AR_2=} , {mse_AR_3}")
```

```python
fig, ax = plt.subplots()

x = ['mean', 'last_value', 'AR(3)']
y = [mse_mean, mse_last, mse_AR]

ax.bar(x, y, width=0.4)
ax.set_xlabel('Methods')
ax.set_ylabel('MSE')
ax.set_ylim(0, 5)

for index, value in enumerate(y):
    plt.text(x=index, y=value+0.25, s=str(round(value, 2)), ha='center')

plt.tight_layout()
```
