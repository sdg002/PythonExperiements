# qqplot

# Prep the data

```python
import numpy as np

```

# normal distribution

```python
# Create a dataset with 100 values that follow a normal distribution
np.random.seed(0)
data = np.random.normal(0, 1, 1000)

# View the first 10 values
print(data[:10])
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Create a Q-Q plot with a 45-degree line added to the plot
fig = sm.qqplot(data, line='45')
plt.show()
```

# uniform distribution

```python
# Create a dataset of 100 uniformly distributed values
data_uniform = np.random.uniform(0, 1, 1000)

# Generate a Q-Q plot for the dataset
fig_uniform = sm.qqplot(data_uniform, line='45')
plt.show()
```

## Without the line

```python
data_uniform = np.random.uniform(0, 1, 1000)

# Generate a Q-Q plot for the dataset
fig_uniform = sm.qqplot(data_uniform)
plt.show()
```

# Plot with itself

```python
data_uniform = np.random.uniform(0, 1, 1000)

# Generate a Q-Q plot for the dataset
fig_uniform = sm.qqplot(data_uniform)
plt.show()
```
