# 6.3 Identifying a stationary ARMA process

```python
np.random.seed(42)

ar1 = np.array([1, -0.33])
ma1 = np.array([1, 0.9])

ARMA_1_1 = ArmaProcess(ar1, ma1).generate_sample(nsample=1000)
ARMA_1_1[:10]
```

```python
plt.figure(figsize=PLOT_SIZE)
plt.title("Simulated ARMA process")
plt.grid()
plt.plot(ARMA_1_1)
```

# ACF plot

## Lags 30

```python
plot_acf(ARMA_1_1, lags=30);
plt.title("After lag 2, the coeff falls. Notice the sinusoidal pattern (lags=30)")
plt.tight_layout()
```

## Lags 20

```python

plot_acf(ARMA_1_1, lags=20);
plt.title("After lag 2, the coeff falls. Notice the sinusoidal pattern (lags=20)")
plt.tight_layout()
```
